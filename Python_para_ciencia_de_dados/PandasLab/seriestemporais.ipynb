```python

import pandas as pd
from matplotlib import pyplot as plt #importando para fazer graficos
from matplotlib import rcParams #usando rcparams para personalizar os graficos 


import seaborn as sns

#funcao para chamar grafico em folha
%matplotlib inline 

rcParams['figure.figsize'] = 5, 4 # escolhando tamanho das figuras
sns.set_style('whitegrid')
```

```python
caminho = 'Spstore.csv'
df = pd.read_csv(caminho,index_col='Order Date', parse_dates=True, encoding='ISO-8859-1' )
#encoding diferente leva ao erro, precisa especificar
df.head()
```

```python
df2 = df.sample(n=100, random_state=25, axis=0)
plt.xlabel('Order data')
plt.ylabel('Order Quantily')
plt.title('Vendas')
df2['Order Quantity'].plot()
```
